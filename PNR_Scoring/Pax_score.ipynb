```python
import pandas as pd
import numpy as np
```

```python
pnr_path = '..\PNRB-ZZ-20231208_062017.csv'
pax_path = '..\PNRP-ZZ-20231208_111136.csv'
```

```python
pnr = pd.read_csv(pnr_path)
pnr_df = pd.DataFrame(pnr)
print(pnr_df)
```

```python
pax = pd.read_csv(pax_path)
pax_df = pd.DataFrame(pax)
pax_df['SSR_CODE_CD1'] = pax_df['SSR_CODE_CD1'].fillna(0)
pax_df['TierLevel'] = pax_df['TierLevel'].fillna(0)
print(pax_df)

```

```python
departure_list = pnr_df['ORIG_CD'].to_list()
arrival_list = pnr_df['DEST_CD'].to_list()

city_pairs = [tuple([departure_list[i], arrival_list[i]]) for i in range(len(departure_list))]

fltnum_list = pnr_df["FLT_NUM"].to_list()

citypair_dict = {fltnum_list[i] : city_pairs[i] for i in range(len(city_pairs))}
print(citypair_dict)
```

```python
recloc_list = pnr_df["RECLOC"].to_list()
recloc_unique = list(set(recloc_list))

count = 0
connection = []

for i in recloc_unique:
    for j in recloc_list:
        if i == j:
            count += 1
    connection.append([i,count])
    count = 0

print(dict(connection))
```

```python
paxcnt_wt = 50
connect_wt = 100
grp_threshold = 7
grp_wt = 500
class_wt = {'FirstClass' : 2000, 'BusinessClass' : 1750, 'PremiumEconomyClass' : 1500, 'EconomyClass' : 1250}
```

```python
def get_recloc_score(df):
    recloc_sclist = []
    recloc_score = 0
    for i in range(len(df['RECLOC'])):
        recloc_score += df['PAX_CNT'][i]*paxcnt_wt + class_wt[df['COS_CD'][i]]
        if df['PAX_CNT'][i] >= grp_threshold:
            recloc_score += grp_wt
        recloc_sclist.append(recloc_score)
        recloc_score = 0

    df['RECLOC_SCORE'] = recloc_sclist

    return df    



        
    

```

```python
print(get_recloc_score(pnr_df))
```

```python
recloc_dict = {pnr_df['RECLOC'][i] : pnr_df['RECLOC_SCORE'][i] for i in range(len(pnr_df['RECLOC']))}
print(recloc_dict)
```

```python
loyalty_dict = {'Platinum' : 1800, 'Gold' : 1600, 'Silver' : 1500, 0 : 0}
ssr_wt = 200

```

```python
def get_pnr_score(df):
    pnr_sclist = []
    pnr_score = 0
    for i in range(len(df['RECLOC'])):
        pnr_score += recloc_dict[df['RECLOC'][i]] + loyalty_dict[df['TierLevel'][i]]
        if df['SSR_CODE_CD1'][i] != 0:
            pnr_score += ssr_wt
        
        pnr_sclist.append(pnr_score)
        pnr_score = 0

    df['PNR_SCORE'] = pnr_sclist

    return df    
    
```

```python
print(get_pnr_score(pax_df))
```
