```python
import pandas as pd
```

```python
excel_path = 'Mphasis - Data boiled down.xlsx'
sheet_name_1 = 'Schedule Attributes'
sheet_name_2 = 'Inventory Attributes'
```

```python
sch = pd.read_excel(excel_path, sheet_name = sheet_name_1)
inv = pd.read_excel(excel_path, sheet_name = sheet_name_2)
```

```python
df1 = pd.DataFrame(sch)
df2 = pd.DataFrame(inv)
```

```python
df1['DepartureTime'] = pd.to_datetime(df1['DepartureTime'], format='%H:%M:%S')
df1['ArrivalTime'] = pd.to_datetime(df1['ArrivalTime'], format='%H:%M:%S')

df2['DepartureTime'] = pd.to_datetime(df2['DepartureTime'], format='%H:%M:%S')
df2['ArrivalTime'] = pd.to_datetime(df2['ArrivalTime'], format='%H:%M:%S')
```

```python
#Importing departure and arrival time from Schedule Data to Inventory Data
i=j=0
while i < len(df1) and j < len(df2):
    if(df1.loc[i, 'ScheduleID'] == df2.loc[j, 'ScheduleId']):
        df2.loc[j, 'DepartureTime'] = df1.loc[i, 'DepartureTime']
        df2.loc[j, 'ArrivalTime'] = df1.loc[i, 'ArrivalTime']
        j = j + 1
    else:
        i = i + 1
```

```python
df2.head()
```

```python
df2['DepartureTime'] = pd.to_datetime(df2['DepartureTime'], format='%H:%M:%S')
df2['DepartureTime'] = df2['DepartureTime'].dt.time

df2['ArrivalTime'] = pd.to_datetime(df2['ArrivalTime'], format='%H:%M:%S')
df2['ArrivalTime'] = df2['ArrivalTime'].dt.time
```

```python
df2.head()
```

```python
#Combining Date and Time into a single column for easy time manipulation in Python
df2['Departure'] = 0
df2['Arrival'] = 0

df2['Departure'] = pd.to_datetime(df2['DepartureDate'].astype(str) + ' ' + df2['DepartureTime'].astype(str), format='%Y-%m-%d %H:%M:%S')
df2['Arrival'] = pd.to_datetime(df2['ArrivalDate'].astype(str) + ' ' + df2['ArrivalTime'].astype(str), format='%Y-%m-%d %H:%M:%S') 
```

```python
df2.head()
```

```python
df2.drop(['DepartureDate', 'ArrivalDate','DepartureTime', 'ArrivalTime'], axis=1, inplace=True )
```

```python
col1 = df2.pop('Departure')
df2.insert(2, 'Departure', col1)

col2 = df2.pop('Arrival')
df2.insert(3, 'Arrival', col2)
```
